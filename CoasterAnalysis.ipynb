## Data Preparation

1. Dropping irrelevant columns & rows
2. Identifying duplicated columns

```python
df = df[['coaster_name', 
    # 'Length', 'Speed', 
    'Location', 'Status', 
    # 'Opening date',
      #  'Type'
    'Manufacturer',
    # 'Height restriction', 'Model', 'Height',
      #  'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
      #  'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
      #  'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
      #  'Track layout', 'Fastrack available', 'Soft opening date.1',
      #  'Closing date', 
    # Opened',
    #  'Replaced by', 'Website',
      #  'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      #  'Single rider line available', 'Restraint Style',
      #  'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
    # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
    # 'height_value', 'height_unit', 
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.info()
```

```python
df['opening_date_clean'] = pd.to_datetime(df['opening_date_clean'])
```

```python
df.columns
```

```python
df = df.rename(columns={'coaster_name': 'Coaster_Name',
                   'year_introduced' : 'Year_Introduced',
                   'latitude': 'Latitude',
                   'longitude':'Longitude',
                   'opening_date_clean':'Opening_Date',
                   'speed_mph':'Speed_MPH',
                   'height_ft':'Height_Ft',
                   'Inversion_clean':'Inversion',
                   'Gforce_clean':'Gforce',
                   })
```

```python
df.head()
ax = df['Year_Introduced'].value_counts().head(10) \
.plot(kind = 'bar', title='Top 10 years Coaster Introduced');
ax.set_xlabel('Years')
ax.set_ylabel('Counts');
```

```python
ax = df['Speed_MPH'].plot(kind='hist', bins=20, title='Coaster Speed MPH')
ax.set_xlabel('Speed mph');
```

```python
ax = df['Speed_MPH'].plot(kind='kde', title= 'Coaster Speed mph')
ax.set_xlabel('Speed MPH');
```

```python
operating = df.query('Status == "Operating"')
ax = operating['Type_Main'].value_counts(normalize=True) \
.plot(kind='pie', title='Operational Coaster Material type', shadow=True, autopct='%1.1f%%')
plt.show()
```
